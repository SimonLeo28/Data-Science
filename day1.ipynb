```python
# Data Structures - List , Tuple , etc...
# List - collection of any data type; []; mutable(can be changed, can have duplicates)
list = ['apple', 'banana', 'cherry']
list
print(dir(list))

#append
list.append('orange')
list

#index
list.index('banana')

#Reversing string
list[::-1]

#pop and remove
list.pop(-1)
list.remove('banana')
list.reverse()
list.sort()
list.clear() #Empties the list
list
```

```python
#tuple - immutable | () | all datatypes
tuple = ('Simon',21,'Banglore',43,True)
tuple
print(dir(tuple))
```

```python
#Conditional Statements - if , else and elif
x = 9
if x >= 10:
    print("True") #there is no alternative output if the value s not >=10
else:
    print("false")
```

```python
x
```

```python
marks = 59
if marks >=90:
    print("A+")
elif marks >=70 and marks <=89:
    print("A")
elif marks>=60 and marks <=69:
    print("B")
else:
    print("C")
```

```python
#even using for loop
for a in range(0,11,2):
    print(a)
```

```python
#odd using for loop
for b in range(1,11,2):
    print(b)
```

```python
#While loop
count = 0
while count <= 5:
    print(count)
    count += 1 #incremental value , if missed will go to infinite loop
```

```python
count = 5
while count >=0:
    print(count)
    count -= 1
```

```python
#break
for i in range(5):
    if i == 3:
        break
    print(i)
```

```python
#continue
for i in range(5):
    if i == 3:
        continue
    print(i)
```
