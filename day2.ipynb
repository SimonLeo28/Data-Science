```python
# REGRESSION
# importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#load dataset
data = pd.read_csv('appliance_energy.csv')
data.head
```

```python
data.head()
```

```python
# Missing values
data.isnull().sum()
```

```python
# Feature and Target
# Features are independent variables
# Targets are dependent variables

# 1 Feature and 1 Target is called simple linear regression
```

```python
X = data[['Temperature (Â°C)']]
y = data['Energy Consumption (kWh)']
```

```python
# Feature Train , Feature Test , Target Train , Target Test
X_train , X_test , y_train , y_test = train_test_split(
    X,y,train_size=0.8, random_state=42
)
#train_test_split(input , output, train_size, random_state)]
#train_test_split is used to split the dataset into training and testing sets. It takes the input features, target variable, train size, and random state as parameters. The train size parameter specifies the proportion of the dataset to include in the training set, while the random state parameter ensures reproducibility of the results by controlling the randomness of the split.
```

```python
# Instantiating the model
model = LinearRegression()
```

```python
#training the model
model.fit(X_train,y_train)
```

```python
# predictions   => X_test is for prediction and y_test is for accuracy
y_pred = model.predict(X_test)
y_pred
```

```python
# metrics - mse & r2 => mean squared error and r2 score
mse = mean_squared_error(y_test,y_pred)
''' mean squared error is a measure of the average squared difference between the predicted values and the actual values.
It quantifies the overall error of the model's predictions, with lower values indicating better performance.'''
mse
```

```python
r2 = r2_score(y_test,y_pred)
```

```python
r2
```

```python
''' R Squared Score is a statistical measure that represents the proportion of the variance in the dependent variable
that is predictable from the independent variable(s). It ranges from 0 to 1, where a value of 1
indicates that the model perfectly explains the variance in the target variable, while a value
of 0 indicates that the model does not explain any of the variance. A higher R Squared Score
indicates a better fit of the model to the data.'''
print(f"R Squared Score: {r2*100:.2f}%")
```

```python
import joblib
joblib.dump(model, 'Simple Linear Regression Model.pkl')
```

```python
plt.scatter(X_test, y_test, color='blue', label='Actual Data')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Regression    Line')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Energy Consumption (kWh)')
plt.title('Simple Linear Regression: Temperature vs Energy Consumption')
plt.legend()
plt.show()
```

```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
gt = pd.read_csv('green_tech_data.csv')
gt.head()
```

```python
gt.isnull().sum()
```

```python
# Feature and Target Selection
gt.columns
```

```python
x = gt[['carbon_emissions', 'energy_output', 'renewability_index','cost_efficiency']]
y = gt['sustainability']
```

```python
xtr , xte , ytr , yte = train_test_split(
    x,y,test_size=0.2, random_state=11
)
```

```python
# Instantiating the model
lr_model = LogisticRegression();

#Training the model
lr_model.fit(xtr,ytr)
```

```python
# Prediction

lr_predict = lr_model.predict(xte)
lr_predict
```

```python
# Evaluation - metrics - accuracy , cr , cm
acc = accuracy_score(yte,lr_predict)
print(f"Accuracy Score: {acc*100:.2f}%")
```

```python
# Confusion Matrix
cm = confusion_matrix(yte, lr_predict)
cm
```

```python
# Classification Report
cr = classification_report(yte, lr_predict)
print(cr)
```
